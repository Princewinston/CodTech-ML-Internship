This project demonstrates sentiment analysis on customer reviews using Natural Language Processing (NLP), TF-IDF Vectorization, and Logistic Regression.

```python
# Import required libraries

import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Load Customer Reviews dataset

df=pd.read_csv("customer_reviews.tsv",sep='\t')
```

```python
# Separate features and target

X=df['Review']
y=df['Liked']
```

```python
# Convert text data into numerical vectors using TF-IDF

vectorizer = TfidfVectorizer()

X_vectorized = vectorizer.fit_transform(X)
```

```python
# Split dataset into training and testing data

X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized,
    y,
    test_size=0.4,
    random_state=42
)
```

```python
# Create Logistic Regression model & Train the model

model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
# Predict sentiments for test data

y_pred = model.predict(X_test)
```

```python
# Calculate model accuracy


accuracy = accuracy_score(y_test, y_pred)

print("\nModel Accuracy:", accuracy)
```

```python
# Test model using custom review

sample_review = ["This product is very excellent"]

# Convert custom review into vector format

sample_vector = vectorizer.transform(sample_review)

# Predict sentiment

prediction = model.predict(sample_vector)

# Convert numeric prediction into text label

if prediction[0] == 1:
    sentiment = "Positive"
else:
    sentiment = "Negative"

print("\nCustom Review Prediction:")
print("Review:", sample_review[0])
print("Predicted Sentiment:", sentiment)
```

# Analysis and Conclusion

The sentiment analysis model was successfully implemented using TF-IDF Vectorization and Logistic Regression.

The model was trained to classify customer reviews as Positive or Negative based on text data.

TF-IDF Vectorization converted text reviews into numerical format so that the machine learning model could process them effectively.

The Logistic Regression model learned patterns from customer reviews and predicted sentiments with good accuracy.
